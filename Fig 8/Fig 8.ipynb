```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
from matplotlib.ticker import AutoMinorLocator

# 
file_path = 'Density2.xlsx'
data = pd.read_excel(file_path)
x = data['WS']

# bin21.527.51
custom_bins = np.arange(0.05, 0.65, 0.1)  # [21.5, 22.5, ..., 27.5]

# 
plt.figure(figsize=(8, 6))

# bins
hist = plt.hist(x, bins=custom_bins, density=True,
                alpha=0.5, color='#D7EBAD',
                edgecolor='#9ACD32',
                label='Temperature Distribution')


# 
plt.xlabel('Air velocity (m/s)', fontsize=20, fontname='Times New Roman')
plt.ylabel('Probability Density', fontsize=20, fontname='Times New Roman')
plt.xticks(custom_bins,  # xbin
           labels=[f'{b:.1f}' for b in custom_bins],  # 
           fontsize=20, fontname='Times New Roman')
plt.yticks(np.arange(0, 10.1, 2), fontsize=20, fontname='Times New Roman')

# 
ax.set_xlim(0, 0.6)
ax.set_ylim(0, 10)

# 
ax.tick_params(axis='x', pad=10)  # pad
ax.tick_params(axis='y', pad=10)  # pad

# 
ax = plt.gca()
ax.xaxis.set_minor_locator(AutoMinorLocator(2))
ax.yaxis.set_minor_locator(AutoMinorLocator(2))
ax.tick_params(axis='both', which='minor', length=4)

# 
ax.tick_params(axis='x', which='minor', direction='in', length=6)
ax.tick_params(axis='y', which='minor', direction='in', length=6)

# 
ax.tick_params(axis='x', which='major', direction='in', length=10)
ax.tick_params(axis='y', which='major', direction='in', length=10)

# bin
for bin_edge in custom_bins:
    plt.axvline(bin_edge, color='gray', linestyle=':', alpha=0.3, linewidth=0.8)

# 
optimal_temp = 0.5
plt.axvline(x=optimal_temp, color='red', linestyle='--', linewidth=1.5, alpha=0.8)
plt.text(x=optimal_temp-0.2, y=plt.ylim()[1]*0.9,  # 0.190%
         s='Optimal air velocity', 
         color='red', 
         fontsize=20, 
         fontname='Times New Roman',
         va='center')

# y5
ax = plt.gca()
current_yticks = ax.get_yticks()
ax.set_yticks(current_yticks)
ax.set_yticklabels([f"{(y*0.1):.1f}" for y in current_yticks])  # 

# 
plt.tight_layout()
plt.show()
```
